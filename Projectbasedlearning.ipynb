```python
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_excel(r"C:\Users\Sujal\Documents\Realestate.xlsx")
df
```

```python
def graph(x,y):
    dt=df.iloc[x,5:10]
    dt.plot()
    plt.title(y)
    plt.xlabel('Year')
    plt.ylabel('Rates')
    plt.show()
    
```

```python
def bar(x,y):
    dt=df.iloc[x,5:10]
    dt.plot.bar()
    plt.title(y)
    plt.xlabel('Year')
    plt.ylabel('Rates')
    plt.show()
    
```

```python
print("Enter Property Type:")
print("\n")

while(1):
    print("Select: \n1.Flat \n2.Land \n3.Office\n4.Exit")
    p=int(input("Enter Choice(1,2,3,4):"))
    if p==1:
        dv=df[df['Property']=='Flat']
        display(dv)
        print("\n")
    
        while(1):
            print("Enter Size of property")
            print("Select:\n1.1 bhk\n2.2 bhk\n3.2.5 bhk\n4.3 bhk\n5. All Data\n6.Exit")
            a=int(input("Enter choice(1,2,3,4,5,6):"))
            if a==1:
                graph(0,"1 BHK")
                bar(0,"1 BHK")
            elif a==2:
                print("\nNo Property Available\n")
            elif a==3:
                graph(5,"2.5 BHK")
                bar(5,"2.5 BHK")
            elif a==4:
                graph(6,"3 BHK")
                bar(6,"3 BHK")
            elif a==5:
                dt=df.iloc[0:9,4:10]
                dt=dt.set_index('Size')
                dt=dt.loc[['1bhk','2.5bhk','3bhk'],[2019,2020,2021,2022,2023]]
                dt=dt.T
                dt.plot()
                plt.title("Flats")
                plt.xlabel('Year')
                plt.ylabel('Rates')
                plt.show()
            elif a==6:
                break;
        
    elif p==2:
        dv=df[df['Property']=='Land']
        display(dv)
        print("\n")
        while(1):
            print("Enter Size of property")
            print("Select:\n1.1 Hector\n2.2 Hector\n3.3 Hector\n4.4 Hector\n5.All Data\n6.Exit")
            a=int(input("Enter choice(1,2,3,4,5,6):"))
            if a==1:
                graph(2,"1 Hector")
                bar(2,"1 Hector")
            elif a==2:
                print("\nNo property Available\n")
            elif a==3:
                graph(3,"3 Hector")
                bar(3,"3 Hector")
            elif a==4:
                graph(1,"4 Hector")
                bar(1,"4 Hector")
            elif a==5:
                dt=df.iloc[0:9,4:10]
                dt=dt.set_index('Size')
                dt=dt.loc[['1 Hector','3 Hector','4 Hector'],[2019,2020,2021,2022,2023]]
                dt=dt.T
                dt.plot()
                plt.title("Land")
                plt.xlabel('Year')
                plt.ylabel('Rates')
                plt.show()
            elif a==6:
                break;
     
    
    elif p==3:
        dv=df[df['Property']=='Office']
        display(dv)
        print("\n")
        
        while(1):
            print("Enter Size of property")
            print("Select:\n1.1000 Sq Foot\n2.1500 Sq Foot\n3.1200 Sq Foot\n4.1700 Sq Foot\n5.All Data\n6.Exit")
            a=int(input("Enter choice(1,2,3,4,5,6):"))
            if a==1:
                graph(4,"1000 Sq Foot")
                bar(4,"1000 Sq Foot")
            elif a==2:
                graph(7,"1500 Sq Foot")
                bar(7,"1500 Sq Foot")
            elif a==3:
                print("\nNo property Available\n")
            elif a==4:
                graph(8,"1700 Sq Foot")
                bar(8,"1700 Sq Foot")
            elif a==5:
                dt=df.iloc[0:9,4:10]
                dt=dt.set_index('Size')
                dt=dt.loc[['1000 Sq.f','1500 Sq.f','1700 Sq.f'],[2019,2020,2021,2022,2023]]
                dt=dt.T
                dt.plot()
                plt.title("Land")
                plt.xlabel('Year')
                plt.ylabel('Rates')
                plt.show()
            elif a==6:
                break;
                
    elif p==4:
        break;
```
